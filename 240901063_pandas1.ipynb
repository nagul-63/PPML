```python
import pandas as pd
#Empty Dataframe
df=pd.DataFrame()
print(df)
```

```python
emp=pd.Series(['Parker','John','Smith','William'])
id=pd.Series([102,107,109,114])
frame={'Emp':emp,'ID':id}
result=pd.DataFrame(frame)
print("\n Series to Dataframe\n")
print(result)
```

```python
print("\n Extracting one column:\n")
print(result['Emp']) 
```

```python
print("\n Adding newcolumn:\n")
result['Age']=pd.Series([35,24,40,38])
print(result)
```

```python
print("\nDeleting one column:\n")
del result['Age']
print(result)
```

```python
print("\n Extracting the thrid row:\n")
print(result.loc[2])
```

```python
print("\n slice rows:\n",result[1:3])
```
